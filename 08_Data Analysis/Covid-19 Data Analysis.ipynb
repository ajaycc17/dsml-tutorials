```python
import pandas as pd

# converting fetched data to dataframe
df = pd.DataFrame(covid.get_data())
df
```

```python
# sorting by number of confirmed cases and copied to a new dataframe
df1 = (df.sort_values(by=['confirmed'], axis=0, ascending=False, inplace=False)).copy()
```

```python
df1
```

```python
# removing continents from individual country's analysis
continents = ['World', 'Asia', 'Africa', 'Europe', 'North America', 'South America', 'Antarctica', '']
df2 = (df1[df1.country.isin(continents) == False]).copy()
```

```python
# taking top 10 countries
df2 = df2[:10]
df2
```

### Top 10 countries affected by Covid-19 (Line graph)

```python
import matplotlib.pyplot as plt
import numpy as np

x = df2['country']
y = df2['confirmed']

# plotting the points in line graph
plt.figure(figsize=(18, 6))
plt.plot(x, y, linewidth = 2, marker='o', markerfacecolor='blue', markersize=4)

# axes labels
plt.xlabel('Country')
plt.ylabel('Cases')

# giving a title to the graph
plt.title('Top 10 countries affected by COVID-19 (Line graph)')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
plt.show()
```

### Top 10 countries affected by Covid-19 (Bar graph)

```python
# in bar graph
plt.figure(figsize=(18, 6))
plt.bar(x, y, width = 0.7)

# axes-labels
plt.xlabel('Country')
plt.ylabel('Cases')

# giving a title to the graph
plt.title('Top 10 countries affected by COVID-19 (Bar graph)')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
plt.show()
```

### Top 10 countries affected by Covid-19 (Multiline graph)

```python
plt.figure(figsize=(18, 6))
x1 = df2['country']
y1 = df2['confirmed']
plt.plot(x1, y1, label = "confirmed")
 
# line 2 points
x2 = df2['country']
y2 = df2['active']
plt.plot(x2, y2, label = "active")

# line 3
x3 = df2['country']
y3 = df2['deaths']
plt.plot(x3, y3, label = "deaths")

# line 4
x4 = df2['country']
y4 = df2['recovered']
plt.plot(x4, y4, label = "recovered")

plt.xlabel('Country')
plt.ylabel('Cases')

# giving a title to the graph
plt.title('Top 10 countries affected by COVID-19 (Multiline graph)')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
 
# show a legend on the plot
plt.legend()
 
# function to show the plot
plt.show()
```

### Top 10 countries affected by Covid-19 (Scatter graph)

```python
# scatter graph
plt.figure(figsize=(18, 6))

# axes labels
plt.xlabel('Country')
plt.ylabel('Cases')

# giving a title to the graph
plt.title('Top 10 countries affected by COVID-19 (Scatter graph)')

plt.scatter(df2['country'],df2['confirmed'])
plt.scatter(df2['country'],df2['deaths'])
plt.scatter(df2['country'],df2['recovered'])
plt.scatter(df2['country'],df2['active'])

plt.xticks(df2['country'], rotation=90)
plt.legend(['confirmed', 'deaths', 'recovered', 'active'])
plt.ticklabel_format(style='plain', axis='y')

plt.grid()
plt.show()
```

### Top 10 countries affected by Covid-19 (Stacked-area graph)

```python
# stacked-area graph
plt.figure(figsize=(18, 6))

# giving a title to the graph
plt.title('Top 10 countries affected by COVID-19 (Stacked-area graph)')

listy = ['confirmed', 'deaths', 'recovered', 'active']
for i in range(10):
    plt.stackplot(listy, [df2.iloc[i][2], df2.iloc[i]
                  [4], df2.iloc[i][6], df2.iloc[i][7]])

plt.legend(list(df2['country']))
plt.ticklabel_format(style='plain', axis='y')

plt.grid()
plt.show()
```

### Top 10 countries affected by Covid-19 (Multiple bar graph with grids)

```python
# multiple bar graph
plt.figure(figsize=(18, 6))

# giving a title to the graph
plt.title('Top 10 countries affected by COVID-19 (Multiple bar graph with grids)')

listy = ['confirmed','deaths','recovered','active']
X_axis = np.arange(len(listy))
offset = -5

for i in range(10):
    plt.bar(X_axis + offset/12,[df2.iloc[i][2],df2.iloc[i][4],df2.iloc[i][6],df2.iloc[i][7]],0.09)
    offset = offset + 1

plt.legend(list(df2['country']))
plt.xticks(X_axis,listy)
plt.ticklabel_format(style='plain', axis='y')

plt.grid()
plt.show()
```
